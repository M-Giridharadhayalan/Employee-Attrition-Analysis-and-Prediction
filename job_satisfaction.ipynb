```python
#Binary Features are Attrition, Gender, Over 18 and Overtime
df['Attrition'] = df['Attrition'].apply(lambda x: 1 if x =='Yes'else 0)
df['Gender'] = df['Gender'].apply(lambda x: 1 if x =='Male'else 0)
df['Over18'] = df['Over18'].apply(lambda x: 1 if x =='Y'else 0)
df['OverTime'] = df['OverTime'].apply(lambda x: 1 if x =='Yes'else 0)
```

```python
df
```

```python
#OneHot Encoding for BusinessTravel', 'Department', 'EducationField' 'JobRole', 'MaritalStatus'

df=df.join(pd.get_dummies(df['BusinessTravel'])).drop('BusinessTravel',axis=1)
df=df.join(pd.get_dummies(df['Department'],prefix=['Department'])).drop('Department',axis=1)
df=df.join(pd.get_dummies(df['EducationField'],prefix=['Education'])).drop('EducationField',axis=1)
df=df.join(pd.get_dummies(df['JobRole'],prefix=['Role'])).drop('JobRole',axis=1)
df=df.join(pd.get_dummies(df['MaritalStatus'],prefix=['Status'])).drop('MaritalStatus',axis=1)
```

```python
df
```

```python
df=df.map(lambda x: 1 if x is True else 0 if x is False else x)
```

```python
df
```

```python
df=df.drop(['EmployeeCount','Over18','StandardHours'],axis=1)
```

```python
sns.countplot(data=df, x='JobSatisfaction',palette='Set2')
plt.title('JobSatisfaction Distribution')
plt.ylabel('Number of Employees')
plt.show()

JobSatisfaction_rate = df['JobSatisfaction'].value_counts(normalize=True) * 100
print("JobSatisfaction Distribution (%):\n", JobSatisfaction_rate.round(2))
```

```python
#Numerical Features vs JobSatisfaction
features = ['Age', 'DistanceFromHome', 'MonthlyIncome', 'TotalWorkingYears', 'YearsAtCompany']

for feature in features:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x='JobSatisfaction', y=feature, data=df,palette='Set2')
    plt.title(f'{feature} vs. JobSatisfaction')
    plt.show()
```

```python
# Chi-squared test: OverTime vs JobSatisfaction
from scipy.stats import chi2_contingency
ct = pd.crosstab(df['OverTime'], df['JobSatisfaction'])
chi2, p, dof, ex = chi2_contingency(ct)
print(f"Chi-squared p-value for OverTime vs JobSatisfaction: {p:.5f}")
```

```python
#MACHINE LEARNING MODELING
#Features&Traget
x = df.drop('JobSatisfaction', axis=1)
y = df['JobSatisfaction']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
ln=LinearRegression()
ln.fit(x_train,y_train)
```

```python
y_pred = ln.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
```

```python
print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
Dt=DecisionTreeRegressor()
Dt.fit(x_train,y_train)
```

```python
y_pred = Dt.predict(x_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
```

```python
print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
```
